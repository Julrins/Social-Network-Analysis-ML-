# Задача:
## Сгенерируйте случайный граф при помощи метода nx.dense_gnm_random_graph() со следующими параметрами: 
### число вершин = 44, число ребер = 428, seed = 68.

```python
# под задание 1
import networkx as nx
from networkx import density
```

```python
# задаем параметры графа по условию
g = nx.dense_gnm_random_graph(44, 428, seed = 68)
g
```

```python
nx.draw(g)
```

# Задача 1.1. Определите следующие значения
## 1. Плотность графа

```python
density = round(density(g), 3)
density
```

## 2. Длину кратчайшего пути от вершины 3 до вершины 38

```python
shortest_path = nx.shortest_path_length(g, source = 3, target = 38)
shortest_path
```

## 3. Количество вершин, входящих в клику наибольшего размера

```python
clique = list(nx.find_cliques(g))
max_clique = max([len(i) for i in clique])
max_clique
```
